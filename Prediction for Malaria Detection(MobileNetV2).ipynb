```python
#Load the model 

from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
import numpy as np


model = load_model('model_mobilenetv2.h5')
```

```python
import os 
import random 
import cv2
import matplotlib.pyplot as plt
%matplotlib inline 

list = ('cell_images/test/Uninfected/', 'cell_images/test/Parasitized/')

path =  random.choice(list) 
files = random.choice(os.listdir(path))
pic = os.path.join(path, files)

pict =  cv2.imread(pic)
plt.imshow(pict)
plt.title(pic)
```

```python
img=image.load_img(pic,target_size=(224,224))
```

```python
x=image.img_to_array(img)
x
```

```python
x.shape
```

```python
x=x/255
```

```python
x=np.expand_dims(x,axis=0)
img_data=preprocess_input(x)
img_data.shape
```

```python
model.predict(img_data)
```

```python
a=np.argmax(model.predict(img_data), axis=1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
if(a==1):
    plt.imshow(img)
    plt.title('Uninfected')
else:
    plt.imshow(img)
    plt.title('Parasited')
```
